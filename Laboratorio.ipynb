# Gráfica ventas del 2020

```python
import matplotlib.pyplot as plt
import seaborn as sns

X = list(data["TiendasVentas"])
Y = list(data["Ventas2020"])
fig, axs = plt.subplots(figsize=(30,10),sharey=True)
axs.bar(X,Y)
plt.xlabel("Tiendas")
plt.ylabel("Ventas")
plt.title("VENTAS 2020")
```

# Gráfica ventas del 2021

```python
X = list(data["TiendasVentas"])
Y = list(data["Ventas2021"])
fig, axs = plt.subplots(figsize=(35,10),sharey=True)
axs.bar(X,Y)
plt.xlabel("Tiendas")
plt.ylabel("Ventas")
plt.title("VENTAS 2021")
```
